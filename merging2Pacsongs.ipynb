```python
### Creating a single dictionary for all 2Pac lyrics ###

os.listdir(path = '2Pac_songs') #This function creates a list of all files in a directory

all2Pacsongs = {}
all2Pacsongs["songs"] = []

for song in os.listdir(path = '2Pac_songs'):
    with open('./2Pac_songs/' + song) as song_file:
        data = json.load(song_file)
        all2Pacsongs["songs"].extend(data["songs"]) #For each of the dictionaries contained in the songs, append the empty dictionary with lyrics

print(all2Pacsongs['songs'][3])
```

```python
all2PacALBUMsongs = {}
all2PacALBUMsongs['songs'] = []

albums_2Pac= ['2Pacalypse Now', 'Strictly 4 My N.I.G.G.A.Z...', 'Me Against the World', 'All Eyez on Me', 'The Don Killuminati: The 7 Day Theory', 'R U Still Down? (Remember Me)', 'Still I Rise', 'Until the End of Time', 'Better Dayz', 'Loyal to the Game', "Pac's Life"]

for song in all2Pacsongs['songs']:
    if song['album'] in albums_2Pac:
        all2PacALBUMsongs['songs'].append(song)
print(all2PacALBUMsongs)
```

```python
with open('all2PacALBUMsongs.json', 'w') as fp:
    json.dump(all2PacALBUMsongs, fp)
```
